```python
from keras.models import load_model
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

%matplotlib inline
```

```python
model = load_model('vgg19_1_chicken.h5')
```

```python
images = ['home.jpeg', 'home_1.jpeg']
classes = {0: 'aloo beans', 1: 'apple', 2: 'banana', 3: 'besan ladoo', 4: 'bhindi sabji', 5: 'chicken curry'}
```

```python
for image in images:
    image = 'test/' + image
    pil_im = Image.open(image, 'r')
    plt.imshow(np.asarray(pil_im))
    plt.show()
    im = cv2.imread(image)
    im = cv2.resize(cv2.cvtColor(im, cv2.COLOR_BGR2RGB), (128, 128)).astype(np.float32) / 255.0
    im = np.expand_dims(im, axis =0)
    print('This is', classes[np.argmax(model.predict(im))])
```
